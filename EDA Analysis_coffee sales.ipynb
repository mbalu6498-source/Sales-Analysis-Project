```python
print(df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['datetime'] = pd.to_datetime(df['datetime'])
df.dtypes
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['weekday'] = df['datetime'].dt.day_name()
```

```python
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(y='coffee_name', data=df, order=df['coffee_name'].value_counts().index)
plt.title("Most Popular Coffee Types")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['money'], kde=True)
plt.title("Distribution of Transaction Amounts")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='money', y='coffee_name', data=df, estimator=np.mean)
plt.title("Average Money Spent by Coffee Type")
plt.show()
```

```python
daily_sales = df.groupby('date')['money'].sum()

plt.figure(figsize=(10,5))
plt.plot(daily_sales.index, daily_sales.values)
plt.title("Daily Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='hour', data=df)
plt.title("Number of Sales by Hour of Day")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='weekday', y='money', data=df, estimator=np.sum, order=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])
plt.title("Total Sales by Day of Week")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='cash_type', data=df)
plt.title("Transaction Count by Payment Type")
plt.show()

```

```python
print("Total Sales:", round(df['money'].sum(), 2))
print("Average Transaction:", round(df['money'].mean(), 2))
print("Most Popular Coffee:", df['coffee_name'].mode()[0])
print("Most Active Hour:", df['hour'].mode()[0])
```
